# Case Study

```python
import matplotlib.pyplot as plt
import numpy as np

x_data = np.linspace(-5, 5, 100)
y_data = np.multiply(-2, x_data) + -3 + np.random.normal(loc=0, scale=0.3, size=(100,))

plt.figure(figsize=(16, 9))
plt.scatter(x_data, y_data, s=10, color='g')
plt.title('y=2x+3')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

w = np.linspace(-5, 5, 100)
b = np.linspace(-5, 5, 100)
b = np.reshape(b, (-1, 1))
w = np.reshape(w, (1, -1))
# w_mesh & b_mesh is shape of (100, 100)
# w_mesh, b_mesh = np.meshgrid(w, b)
x_data = np.reshape(x_data,(-1, 1))

c = np.square(w * x_data + b - y_data) / 100

figure = plt.figure(figsize=(16, 9))
axes = Axes3D(figure)
axes.set_xlabel('w')
axes.set_ylabel('b')
axes.plot_surface(w, b, c, rstride=1, cstride=1, cmap='rainbow')
```
