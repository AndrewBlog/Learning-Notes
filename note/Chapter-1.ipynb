# Case Study

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

data_count = 100

x_data = np.linspace(-20, 20, data_count)
y_data = np.multiply(2, x_data) + 3 + np.random.normal(loc=0, scale=1.0, size=(data_count,))

plt.figure(figsize=(16, 9))
plt.scatter(x_data, y_data, s=10, color='g')
plt.title('y=2x+3')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

w_sample = np.linspace(-10, 10, data_count).reshape((-1, 1))
b_sample = np.linspace(-10, 10, data_count).reshape((-1, 1))

x_data = x_data.reshape((-1, 1))
y_data = y_data.reshape((-1, 1))

w_mesh, b_mesh = np.meshgrid(w_sample, b_sample)

loss = np.square(w_mesh * x_data.T + b_mesh - y_data.T) / data_count * 0.5

figure = plt.figure(figsize=(16, 9))
axes = Axes3D(figure)
axes.set_xlabel('w')
axes.set_ylabel('b')
axes.plot_surface(w_mesh, b_mesh, loss, cmap='rainbow')
```

```python
w = np.random.rand()
b = np.random.rand()
y_predict = w * x_data + b

plt.figure(figsize=(16, 9))
plt.scatter(x_data, y_data, s=10, color='g')
plt.plot(x_data, y_predict)
plt.title('y=2x+3')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
for iteration in range(3000):
    y_predict = w * x_data + b
    grad_w = np.mean((y_predict - y_data) * x_data)
    grad_b = np.mean((y_predict - y_data))
    w -= 0.003 * grad_w
    b -= 0.003 * grad_b

y_predict = w * x_data + b

plt.figure(figsize=(16, 9))
plt.scatter(x_data, y_data, s=10, color='g')
plt.plot(x_data, y_predict)
plt.title('y=2x+3')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
