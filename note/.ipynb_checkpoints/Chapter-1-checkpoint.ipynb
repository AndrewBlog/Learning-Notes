# Case Study

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

data_count = 100

x_data = np.linspace(-20, 20, data_count)
y_data = np.multiply(2, x_data) + 3 + np.random.normal(loc=0, scale=1.0, size=(data_count,))

plt.figure(figsize=(16, 9))
plt.scatter(x_data, y_data, s=10, color='g')
plt.title('y=2x+3')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

w_sample = np.linspace(-10, 10, data_count).reshape((-1, 1))
b_sample = np.linspace(-10, 10, data_count).reshape((-1, 1))

x_data = x_data.reshape((-1, 1))
y_data = y_data.reshape((-1, 1))

w_mesh, b_mesh = np.meshgrid(w_sample, b_sample)

loss = np.square(w_mesh * x_data.T + b_mesh - y_data.T) / data_count * 0.5

figure = plt.figure(figsize=(16, 9))
axes = Axes3D(figure)
axes.set_xlabel('w')
axes.set_ylabel('b')
axes.plot_surface(w_mesh, b_mesh, loss, cmap='rainbow')
```

```python
w = np.random.normal(loc=0, scale=0.3)
b = np.random.normal(loc=0, scale=0.3)
y = w * x_data + b

plt.figure(figsize=(16, 9))
plt.scatter(x_data, y_data, s=10, color='g')
plt.plot(x_data, y)
plt.title('y=2x+3')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
grad_w = (1 / data_count) * np.sum(w * x_data + b - y_data) * np.sum(x_data)
grad_b = (1 / data_count) * np.sum(w * x_data + b - y_data)

for iteration in range(300):
    w = w - 0.03 * grad_w
    b = b - 0.03 * grad_b
    grad_w = (1 / data_count) * np.sum(w * x_data + b - y_data) * np.sum(x_data)
    grad_b = (1 / data_count) * np.sum(w * x_data + b - y_data)

    print(np.mean(np.square(w * x_data + b - y)))

y = w * x_data + b

plt.figure(figsize=(16, 9))
plt.scatter(x_data, y_data, s=10, color='g')
plt.plot(x_data, y)
plt.title('y=0.5x+3')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
